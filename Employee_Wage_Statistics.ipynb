```python
df['Salary'] = df['Salary'].replace('[\$,]', '', regex=True).astype(float)
```

# Fictitious Business

# Analytical Questions

# Gender distribution within different positions

```python
gender_counts = df.groupby(['Position', 'Gender']).size().unstack().fillna(0)
```

```python
gender_counts
```

```python
gender_counts.plot(kind='bar', stacked=True, figsize=(14, 7))
plt.title('Gender Distribution by Position')
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.xticks(rotation=90)
plt.legend(title='Gender')
plt.show()
```

The gender distribution varies significantly across different positions. Positions such as Administrative Assistant and Accountant I have more females, while technical and managerial positions like Network Engineer and Production Manager have a higher number of males. This suggests a potential gender imbalance in certain roles.Here IT manager positions were also dominated by men.Excecutive positions are dominated by females

# Median salary based on age segments

```python
age_bins = [20, 30, 40, 50, 60, 70]
age_labels = ['20-30', '31-40', '41-50', '51-60', '61-70']
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)
median_salary_age_group = df.groupby('Age_Group')['Salary'].median()
median_salary_age_group
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=median_salary_age_group.index, y=median_salary_age_group.values)
plt.title('Median Salary by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Salary in USD')
plt.show()
```

Median salaries for the age group 61-70 are significantly higher than other age groups. The 20-30 age group has a higher median salary compared to 31-50 age groups, indicating higher entry-level pay for newer roles or higher-skilled younger employees.

# Salary Variance by Department (Coefficient of Variation)

```python
department_sal = df.groupby('Department')['Salary'].agg(['mean', 'median', 'std', 'count','max','min'])
department_sal['Coefficient_of_Variation'] = (department_sal['std'] / department_sal['mean']) * 100
department_sal.sort_values('mean', ascending=False)
```

The Executive Office has the highest mean salary.Sales department shows the lowest coefficient of variation, indicating relatively consistent salaries.Production department has a high coefficient of variation, indicating significant inconsistency in salaries within the department.production has highest count of 106 employes.

# Ethnic Diversity by Department

```python
ethnicity_count_departments = df.groupby(['Department', 'Ethnicity']).size().unstack().fillna(0)
ethnicity_count_departments
```

```python
ethnicity_count_departments.plot(kind='bar', stacked=True, figsize=(8,8))
plt.title('Ethnicity Distribution by each Department')
plt.xlabel('Nmae of the Department')
plt.ylabel('Number of Employees')
plt.legend(title='Ethnicity')
plt.show()
```

The production department has the highest diversity with employes from differtent ethinic backgroungs.IT/IS  and sales also shows diverse ethinic distribution.Excecutive office have only whites.Whites dominate all the departments.

# Gender Diversity in Executive Positions

```python
exec_positions = ['President & CEO', 'CIO', 'Director of Operations', 'Director of Sales']
df['Is_Executive'] = df['Position'].isin(exec_positions)

gender_exec_percent = df[df['Is_Executive']].groupby('Gender').size() / df.groupby('Gender').size() * 100
print("Percentage of Each Gender in Executive Positions")
print()
print(gender_exec_percent)


```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=gender_exec_percent.index, y=gender_exec_percent.values, palette="cool")
plt.title('Percentage of Each Gender in Executive Positions')
plt.xlabel('Gender')
plt.ylabel('Percentage (%)')
plt.show()
```

There are only females in excecutive positions and 4.08 percent of females are in excecutive roles.

# Average Tenure by Department

```python
df['Start_date'] = pd.to_datetime(df['Start_date'])
df['Tenure'] = (pd.to_datetime("today") - df['Start_date']).dt.days / 365


```

```python
ten_dep = df.groupby('Department')['Tenure'].mean()
print("Average Tenure based on Department:")
print()
print(ten_dep)
```

```python
plt.figure(figsize=(12, 6))
ten_dep.plot(kind='bar')
plt.title("Average Tenure based on Department")
plt.xlabel("Department")
plt.ylabel("Average Tenure in Years")
plt.show()
```

All the tenures remains in same level over here with slight variations.Excecutive office has higest average tenure indicatiing stable leaders.IT/IS has the lowest average tenures.

# Age Demographics within Departments

```python
age_by_department = df.groupby('Department')['Age'].describe()
print(age_by_department)
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Department', y='Age', data=df)
plt.xticks(rotation=90)
plt.title('Age Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Age')
plt.show()
```

Higest variation of ages lies in sales department.Excecutive office has the highest age .There are some high ages present in production,software engineering and admin offices which are shown as outliers.IT/IS have broader range of age distributions.Most of the mean ages vary in the range of 32-39 except exceutive office with admin offices being the lowest and production being the highest.
