```python
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Dependents'].fillna(df['Dependents'].mode()[0], inplace=True)
df['Self_Employed'].fillna(df['Self_Employed'].mode()[0], inplace=True)
df['Education'].fillna(df['Education'].mode()[0], inplace=True)
```

There are missing values and here i replaced the missing values with the mode to replace the missing values.

```python
df.isnull().sum()*100/len(df)
```

```python
df=df.dropna()
```

```python
df.info()
```

THe above missing values are important and if we fill the values with mean or some other values it will affect the loan status and these are sensitive variables and they vary from 2 to 7 percent and so i dropped 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History' these sensiteve information with null values were dropped

```python
for i in ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']:
  sns.boxplot(df[i])
  plt.title(i)
  plt.show()
```

```python
column = 'ApplicantIncome'
q1 = df[column].quantile(0.25)
q3 = df[column].quantile(0.75)
iqr = q3 - q1
upper_boundary = q3 + 1.5 * iqr
lower_boundary = q1 - 1.5 * iqr
df[(df[column] < lower_boundary) | (df[column] > upper_boundary)][column]
```

```python
column = 'CoapplicantIncome'
q1 = df[column].quantile(0.25)
q3 = df[column].quantile(0.75)
iqr = q3 - q1
upper_boundary = q3 + 1.5 * iqr
lower_boundary = q1 - 1.5 * iqr
df[(df[column] < lower_boundary) | (df[column] > upper_boundary)][column]
```

```python
column = 'LoanAmount'
q1 = df[column].quantile(0.25)
q3 = df[column].quantile(0.75)
iqr = q3 - q1
upper_boundary = q3 + 1.5 * iqr
lower_boundary = q1 - 1.5 * iqr
df[(df[column] < lower_boundary) | (df[column] > upper_boundary)][column]
```

```python

for column in ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']:
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    upper_boundary = q3 + 1.5 * iqr
    lower_boundary = q1 - 1.5 * iqr
    df[column] = df[column].clip(lower=lower_boundary, upper=upper_boundary)
```

```python
for i in ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']:
  sns.boxplot(df[i])
  plt.title(i)
  plt.show()
```

There are outliers present in the columns ApplicantIncome, CoapplicantIncome, LoanAmount and i removed the outliers by replacing the lower outliers with the lower boundary which is  q1 - 1.5 * iqr value and upper values were replaced by q3 + 1.5 * iqr and this removes the outliers completley.

# The relationships within the dataset - using 2 different methods

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

There's a moderate positive correlation  between these two variables Applicant Income and Loan Amount. This suggests that as the applicant's income increases, the loan amount they tend to apply for also increases.

There's a weak positive correlation (0.13) between these two variables Loan Amount and Loan Amount Term. This suggests that there might be a slight tendency for larger loan amounts to have longer terms.

There's a moderate negative correlation (-0.38) between these two variables Applicant Income and Coapplicant Income. This indicates that as the applicant's income increases, the coapplicant's income tends to decrease. This could be due to various reasons, such as the applicant being the primary income earner in the household.




```python
sns.pairplot(df, hue='Loan_Status')
```

Credit History:Credit History appears to be the most critical factor in loan approval. Applicants with a good credit history (Credit_History = 1) have a much higher likelihood of loan approval.

Applicant Income and Coapplicant Income both show a weak relationship with loan status, as loans are approved across a wide range of income levels. However, it seems that applicants with moderate income levels (for both Applicant and Coapplicant) have a slightly higher chance of approval.

Loan Amount does not have a strong effect on loan status, although extremely high loan amounts are slightly less likely to be approved.

Loan Amount Term is not highly predictive of loan status, though the standard term of 360 months is the most common for approved loans.
